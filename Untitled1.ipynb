```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Download Tesla revenue data
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# Find all tables
tables = soup.find_all("table")

# Check if tables were found
if len(tables) == 0:
    print("No tables found on the webpage.")
else:
    print(f"Found {len(tables)} tables.")

# Extract the first table (Tesla revenue)
tesla_table = tables[0]
rows = tesla_table.find_all("tr")

# Extract headers and data
headers = [header.text.strip() for header in rows[0].find_all("th")]
data = [[col.text.strip() for col in row.find_all("td")] for row in rows[1:]]

# Debugging output
print("Headers:", headers)
print("First 5 rows of data:", data[:5])

# Ensure headers match data columns
if len(headers) != len(data[0]):
    print(f"Warning: Header count ({len(headers)}) does not match data column count ({len(data[0])}).")

# Create DataFrame only if headers and data columns match
if len(headers) == len(data[0]):
    tesla_revenue = pd.DataFrame(data, columns=headers)
    print(tesla_revenue.tail())  # Display last 5 rows
else:
    print("Skipping DataFrame creation due to mismatch.")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL containing revenue data
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"

# Fetch HTML content
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# Find all tables
tables = soup.find_all("table")
print(f"Found {len(tables)} tables on the page.")

# Check first few rows of each table
for i, table in enumerate(tables):
    rows = table.find_all("tr")
    headers = [header.text.strip() for header in rows[0].find_all("th")]
    data = [[col.text.strip() for col in row.find_all("td")] for row in rows[1:]]

    print(f"\n🔹 Table {i}:")
    print("Headers:", headers)
    print("First 5 rows of data:", data[:5])

# Manually check which table contains Tesla and GameStop revenue
# Assuming Tesla revenue is in tables[0] and GameStop revenue is in tables[1] (adjust if needed)
tesla_table = tables[0]  # Try changing to tables[1] if wrong
gme_table = tables[1]  # Try changing to tables[0] if wrong

# Extract Tesla revenue
tesla_rows = tesla_table.find_all("tr")
tesla_headers = [header.text.strip() for header in tesla_rows[0].find_all("th")]
tesla_data = [[col.text.strip() for col in row.find_all("td")] for row in tesla_rows[1:]]

# Extract GameStop revenue
gme_rows = gme_table.find_all("tr")
gme_headers = [header.text.strip() for header in gme_rows[0].find_all("th")]
gme_data = [[col.text.strip() for col in row.find_all("td")] for row in gme_rows[1:]]

# Check if headers match data
if len(tesla_headers) == len(tesla_data[0]):
    tesla_revenue = pd.DataFrame(tesla_data, columns=tesla_headers)
    print("\n✅ Tesla Revenue DataFrame Created Successfully")
    print(tesla_revenue.tail())
else:
    print("\n⚠️ Tesla revenue data format mismatch! Check table index.")

if len(gme_headers) == len(gme_data[0]):
    gme_revenue = pd.DataFrame(gme_data, columns=gme_headers)
    print("\n✅ GameStop Revenue DataFrame Created Successfully")
    print(gme_revenue.tail())
else:
    print("\n⚠️ GameStop revenue data format mismatch! Check table index.")
```

```python
# Download GameStop revenue data
url_gme = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
response_gme = requests.get(url_gme)
soup_gme = BeautifulSoup(response_gme.text, "html.parser")

# Extract revenue table
tables_gme = soup_gme.find_all("table")
gme_table = tables_gme[1]  # Assuming GameStop's table is the second one
rows_gme = gme_table.find_all("tr")

# Extract headers and data
headers_gme = [header.text.strip() for header in rows_gme[0].find_all("th")]
data_gme = [[col.text.strip() for col in row.find_all("td")] for row in rows_gme[1:]]

# Create DataFrame
gme_revenue = pd.DataFrame(data_gme, columns=headers_gme)

# Display last five rows
print(gme_revenue.tail())

# Save to CSV (optional)
gme_revenue.to_csv("gme_revenue_data.csv", index=False)
```
